# **4. Animation from Scratch**

# 4.1 Adjust the data frame

```python
## Be careful !!! YOU NEED TO USE AN UNINDEXED DATA FRAME (no con d2 sino con d)

# Example 1
### Rows = years
### columns = country
### values = life

l2 = latam.pivot( index= "year",
                    columns= "country",
                    values =  "life")
l2 = l2.reset_index()
l2['year_str'] = l2['year'].astype(str)
l2.head()
```

## 4.2 Create the frame

```python
# The frame
fig = go.Figure(
    layout = go.Layout(
        updatemenus = [dict(type = 'buttons', direction = 'right', x = 0.9, y = 1.16),],
        xaxis = dict(range = ['2016', '2020'],
                     autorange = False,
                     tickwidth = 2, dtick = 1,
                     title_text = 'Year'),
        yaxis = dict(range = [68,85],
                     autorange = False,
                     title_text = 'Life'),
        title = 'Life Expectancy in principal LATAM countries',
        title_font_size = 30,
        title_x = 0.5
    )
)
fig.show()
```

```python
## Add Traces
init = 1

### Colombia
fig.add_trace(
    go.Scatter(
        x = l2.year[:init],
        y = l2.Colombia[:init],
        name = 'Colombia',
        line = dict(color = 'black'),
        mode = 'lines'
    )
)

### Chile
fig.add_trace(
    go.Scatter(
        x = l2.year[:init],
        y = l2.Chile[:init],
        name = 'Chile',
        line = dict(color = 'red'),
        mode = 'lines'
    )
)

### Ecuador
fig.add_trace(
    go.Scatter(
        x = l2.year[:init],
        y = l2.Ecuador[:init],
        name = 'Ecuador',
        line = dict(color = 'green'),
        mode = 'lines'
    )
)

### Mexico
fig.add_trace(
    go.Scatter(
        x = l2.year[:init],
        y = l2.Mexico[:init],
        name = 'Mexico',
        line = dict(color = 'blue'),
        mode = 'lines'
    )
)

### Paraguay
fig.add_trace(
    go.Scatter(
        x = l2.year[:init],
        y = l2.Paraguay[:init],
        name = 'Paraguay',
        line = dict(color = 'orange'),
        mode = 'lines'
    )
)
```

```python
## Frames
frames = [
    go.Frame(
        data = [
            go.Scatter(x = l2.year[:k], y = l2.Colombia[:k]),
            go.Scatter(x = l2.year[:k], y = l2.Chile[:k]),
            go.Scatter(x = l2.year[:k], y = l2.Ecuador[:k]),
            go.Scatter(x = l2.year[:k], y = l2.Mexico[:k]),
            go.Scatter(x = l2.year[:k], y = l2.Paraguay[:k])
        ]
    )
    for k in range(init, len(l2)+1)
]

## Animation
fig.update(frames=frames)
```

```python
## Play button
fig.update_layout(
    updatemenus = [
        dict(
            buttons = list([
                dict(
                label = 'Play',
                method = 'animate',
                args = [None, {'frame':{'duration':500}}]
                )
            ]
            )
        )
    ]
)
```
