# **1. Libraries**

```python
# Read the file

## Save the path
p = '/content/BASE DE DATOS ACT.xlsx'

## Read and create the dataframe
l =pd.read_excel(p)
l.info()
```

# **3.Cleaning**

## Column names

```python
# Time to time
l.rename(columns = {'Time':'year'}, inplace =True)

# Country Name to country
l.rename(columns = {'Country Name':'country'}, inplace =True)

#Time code to codet
l.rename(columns = {'Time Code':'codet'}, inplace =True)

# Country Code to code
l.rename(columns = {'Country Code':'code'}, inplace =True)

# Life expectancy at birth, total (years)
l.rename(columns = {'Life expectancy at birth, total (years) [SP.DYN.LE00.IN]':'life'}, inplace =True)

# People using safely managed drinking water services (% of population)
l.rename(columns = {'People using safely managed drinking water services (% of population) [SH.H2O.SMDW.ZS]':'water'}, inplace =True)

# Domestic general government health expenditure (% of GDP)
l.rename(columns = {'Domestic general government health expenditure (% of GDP) [SH.XPD.GHED.GD.ZS]':'gov'}, inplace =True)


l.head()
```

# Delete columns

```python
l = l.drop(columns=['codet'])
l.info()
```

## **4. Indexing**

```python
# Duplicate the dataframe
l2 = l
```

```python
# Duplicate year as string / How to create a column
l2['year_str'] = l2['year'].astype(str)
l2.info()
```

```python
# Indexing
idx = pd.MultiIndex.from_product([l2['country'].unique(),range(l2.year.min(), l2.year.max()+1)],
                                 names = ['country','year'])

## Attach the index
l2 = l2.set_index(['country','year'])

l2.head()
```

# **5. Delete Nan ..**

```python
l2.isna()
```

```python
# By the files
l2 = l2.replace('..',np.NaN)
l2.head()
```

```python
# Delete NaN
l2 = l2.dropna()
l2.info
```

```python
# By the flies
l2 = l2.replace("nan",np.NaN)
l2.head()
```

```python
# Delete Nan
l2 = l2.dropna()
l2.info()
```
