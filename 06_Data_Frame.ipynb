# **3.Cleaning**

## Column names

```python
# Time to time
l.rename(columns = {'Time':'year'}, inplace =True)

# Country Name to country
l.rename(columns = {'Country Name':'country'}, inplace =True)

#Time code to codet
l.rename(columns = {'Time Code':'codet'}, inplace =True)

# Country Code to code
l.rename(columns = {'Country Code':'code'}, inplace =True)

# Life expectancy at birth, total (years)
l.rename(columns = {'Life expectancy at birth, total (years)':'life'}, inplace =True)

# People using safely managed drinking water services (% of population)
l.rename(columns = {'People using safely managed drinking water services (% of population)':'water'}, inplace =True)

# Consumption of iodized salt (% of households)
l.rename(columns = {'Consumption of iodized salt (% of households)':'salt'}, inplace =True)

# Domestic general government health expenditure (% of GDP)
l.rename(columns = {'Domestic general government health expenditure (% of GDP)':'gov'}, inplace =True)


l.head()
```
